```python
%load_ext cudf.pandas
import pandas as pd
pd
```

```python
df = pd.read_csv('heart_disease_health_indicators_BRFSS2015.csv')
df
```

```python
from cuml.preprocessing import MaxAbsScaler
import cupy as cp
```

```python
class Normalization(object):
    def __init__(self, dataset):
        self.dataset = dataset.copy().reset_index(drop = True)


    def MaxAbsScaler(self):
        global maxAbsScaler_global
        df_maxAbs_scaler = self.dataset.copy()
        df_maxAbs_scaler.to_numpy()
        X = cp.array(df_maxAbs_scaler)
        transformer = MaxAbsScaler().fit(X)
        df_maxAbs_scaler_transform = transformer.transform(X)
        
        maxAbsScaler_global = pd.DataFrame(df_maxAbs_scaler_transform)
        

    def main(self):
        self.MaxAbsScaler()
        
```

```python
norm = Normalization(df)
```

```python
import time

st = time.time()
norm.main()
et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')
```
